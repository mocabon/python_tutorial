### 시각화 심화

```python
import pandas as pd 
# library & dataset
import seaborn as sns
%matplotlib inline
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")

# 컬럼해더 재정의
customerData.columns = ['CUSTID','AVGPRICE','EMI','DEVICECOUNT','PRODUCTAGE','CUSTTYPE']

# 데이터 VIEW
customerData.head()
```

```python
customerData.corr()
```

```python
tc = customerData.corr()
sns.heatmap(tc, annot = True)
```

```python
g= sns.pairplot(data=customerData, hue = "CUSTTYPE")
# plt.legend(loc='upper left', bbox_to_anchor=(0.0, 1.15), ncol=3) #horizontal legend top
# g.savefig('d:/Test.png', bbox_inches='tight')
```

```python
sns.lmplot( x="DEVICECOUNT", y="PRODUCTAGE",data=customerData, hue="CUSTTYPE", fit_reg=False)
```

```python
colorsmap = {'Big-Screen-lover':'red', 'Sleeping-dog':'blue', 'Early-bird':'green'}
codemap = {'Big-Screen-lover':0, 'Sleeping-dog':1, 'Early-bird':2}
```

```python
customerData["COLOR"] = customerData.CUSTTYPE.apply(lambda x: colorsmap[x])
customerData["CODE"] = customerData.CUSTTYPE.apply(lambda x: codemap[x])
customerData.head()
```

```python
customerData.plot(kind="scatter",x="DEVICECOUNT", y="PRODUCTAGE", c=customerData.COLOR, legend=True)
```
